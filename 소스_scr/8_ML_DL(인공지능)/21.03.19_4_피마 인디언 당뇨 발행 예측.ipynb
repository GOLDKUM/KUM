```python
import numpy as np # csv 파일을 numpy배열로 읽기
import tensorflow.keras.utils as utils #분류분석시 원핫인코딩 시
from tensorflow.keras.models import Sequential #모델 생성
from tensorflow.keras.layers import Dense,Activation
```

# 1.데이터 셋 생성

```python
dataset=np.loadtxt('data/pima-indians-diabetes.csv',delimiter=',')
dataset
```

```python
dataset.shape
```

```python
#학습용
x_train=dataset[:700,:8]
y_train=dataset[:700,8]
#테스트
x_text=dataset[700:,:8]
y_text=dataset[700:,8]
```

```python
x_train.shape,y_train.shape,x_text.shape,y_text.shape
```

```python
# 라벨링 0 -> [1, 0] / 1 -> [0,1]
```

# 2.모델 구성하기

```python
model = Sequential()
#input_dim=8 최종 출력 1 sigmod
model.add(Dense(12,input_dim=8,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

# 3. 학습과정 설정

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

# 4. 학습시키기

```python
hist = model.fit(x_train,y_train,epochs=1500,batch_size=64)
# hist = model.fit(x_train,y_train,epochs=100,batch_size=10,
#                 verbose=2,validation_data=(x_train,y_train))
```

```python
hist.history['loss']
```

# 5.모델평가하기

# go to home  모델의 학습과정 표시하기

```python
import matplotlib.pyplot as plt
fig,loss_ax=plt.subplots()
loss_ax.plot(hist.history['loss'],'y',label='train loss')

acc_ax=loss_ax.twinx()# loss_ax 와 x축을 공유하는 acc_ax 생성
acc_ax.plot(hist.history['accuracy'],'g',label='train acc')
loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuracy')
loss_ax.legend(loc='upper left')
acc_ax.legend(loc='upper left')
plt.show()
```

```python
score = model.evaluate(x_text,y_text)
```

```python
print("loss : %s\n accuracy : %.2f"%(score[0],score[1]))
```
