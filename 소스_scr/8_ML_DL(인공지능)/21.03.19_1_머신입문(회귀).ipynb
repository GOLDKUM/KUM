# 1.1 기존의 프로그램 방식

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 섭씨온도(인풋 데이터)를 받아 화씨온도(target 데이터)로 출력
```

```python
def celsius_to_faherenheit(x):
    return x * 1.8 + 32
```

```python
input_c = int(input('섭씨 온도는 ?'))
print("화씨 온도는",celsius_to_faherenheit(input_c))
```

## 1.2 머신러닝 방식
- 데이터 셋(훈련데이터셋, 검증데이터셋,시험데이터 셋)
- 1. 데이터 확보 및 생성
- 2. 데이터 전처리 :훈련데이터셋, 검증데이터셋 ,시험데이터셋,...

- 3. 모델 구성
- 4. 모델 학습과정
- 5. 모델 학습 시키기
- 6. 모델 평가(시험데이터 셋으로)
- 7. 모델 사용(입력값을 주어서 예측값을 받기)

```python
# 1&2. 데이터 셋
data_C = np.array(range(100))
data_C #입력값
```

```python
data_F=celsius_to_faherenheit(data_C)
data_F # target 값
```

```python
#2. 컴퓨터에게 학습시키기 위해 nirmalize함(전체의 편차 줄이기. 이상적인 편차 0~1)
scaled_data_C=data_C/100.0
scaled_data_F=data_F/100.0
print(scaled_data_C,end='\n\n')
print(scaled_data_F)
```

```python
from tensorflow.keras.models import Sequential# 모델 생성
from tensorflow.keras.layers import Dense#입력값과 출력값
```

# 오류  결과값 이해하기
- loss 와 mae의 뜻이 뭐지 이해  (회귀)

```python
#3. 모델 구성 
model = Sequential()
model.add(Dense(1,input_dim=1))# target데이터(종속변수1개) input_dim=1 :독립변수1개 =입력값 1개

#4. 모델 학습과정 설정
model.compile(loss='mse',optimizer='rmsprop',metrics=['mae'])
print(model.summary())
```

```python
# 학습전 예측
print(model.predict([0]))
```

```python
model.save('model/before_learning.h5')
```

```python
#5 모델 학습시키기
hist = model.fit(scaled_data_C,scaled_data_F,epochs=1000,verbose=2)
```

```python
hist.history.keys()
```

```python
plt.plot(hist.history['loss'])
```

```python
model.predict([0.02])
```

```python
model.save('model/after_learning.h5')
```

```python
plt.scatter(x=scaled_data_C,y=scaled_data_F)
```

## 1.3 노이즈가 있는 데이터로 실습

# random.normal 

```python
# 평균이 0이고 표준편차가 0.1인 데이터 100
np.random.seed(703)
noise=np.array(np.random.normal(0,0.1,100))
noised_scaled_data_F=scaled_data_F+noise
```

```python
plt.scatter(x=scaled_data_C,y=noised_scaled_data_F)
plt.show()
```

```python
# 3. 모델 구성 
model2 = Sequential()
model2.add(Dense(1,input_dim=1))
# 4. 학습과정 설정 
model2.compile(loss='mse',optimizer='rmsprop',metrics=['mae'])
print(model2.summary())
```

```python
# 5. 모델 학습시키기
hist2=model2.fit(scaled_data_C,noised_scaled_data_F,epochs=1000,verbose=2)
```

```python
print(model2.predict(np.array([0.01])))
```

```python
plt.plot(hist2.history['loss'])
```

```python
model2.save('model/noise_after_learning.h5')
```
