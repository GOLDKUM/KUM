```python
#방법2 (2행 2열 서브플롯)많이 이용함!
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(15,15))
axes[0,0].plot(x,np.sin(x),color='k',linewidth=3)
axes[0,0].plot(x,np.cos(x),color='r',linewidth=3)
axes[0,0].plot(x,np.sin(x)*np.cos(x),color='b',linewidth=3)
axes[0,1].plot(x,np.cos(x),'y--')
axes[1,0].plot(x,np.tanh(x),'g')
axes[1,1].plot(x,np.sin(x)*np.cos(x),'r--')
plt.show()
```

```python
# 방법3( 2행 2열 서브플롯 ) for문을 이용해 2번쨰방법을 사용
x= np.arange(0,7,0.01)
x.shape
```

```python
# def sin_cos(x):
#     return np.sin(x)*np.cos(x)
# func_list=[np.sin,np.cos,np.tanh,sin_cos]
func_list=[np.sin,np.cos,np.tanh,lambda x: np.sin(x)*np.cos(x)]
func_list[3](4)
fmt_list=['k:','c--','g','r-.']
```

```python
list(axes.flat)
```

```python
plt.subplot(2,2,3);plt.plot(x,np.tanh(x))
```

```python
fig , axes=plt.subplots(2,2,figsize=(10,6))
#2차원 axes를 list(axes.flat) (# flat 할 경우 1차원으로 나열)이용해서 중첩for문 안 씀
for i, ax in enumerate(axes.flat):
    ax.plot(x,func_list[i](x),fmt_list[i])
```

```python
fig,axes=plt.subplots(ncols=4,figsize=(12,3))
for i, ax in enumerate(axes):
    ax.plot(x,func_list[i](x),fmt_list[i])
```

```python
fig,axes=plt.subplots(nrows=4,figsize=(10,10))
for i,ax in enumerate(axes):
    ax.plot(x,func_list[i](x),fmt_list[i])
```

```python
x=np.linspace(0,10)
y=np.cos(x)
fig,axes =plt.subplots(2,2,figsize=(10,5))
axes[0,0].scatter(x,y,s=5,c='r',marker='o')
axes[0,1].fill(x,y,c='b')
axes[1,1].fill_between(x,y,color='r')
axes[1,1].plot([-1,5],[-1,1],'ro-')
```

```python
fig,axes=plt.subplots(2,2)
axes[0,0].bar([1,2,3],[3,4,5],color='r')
axes[0,0].plot([1,2,3],[3,4,5],'k')
axes[0,1].bar([0.5,1,2.5],[0,1,2])
axes[0,1].plot([0.5,1,2.5],[0,1,2],'r')
axes[1,0].axvline(0.5)
axes[1,1].axhline(0.75)
plt.show()
```

**4)축과 눈금**


```python
import numpy as np
x= np.linspace(0,10,100)
y= np.cos(x)
fig,axes=plt.subplots(2,2,figsize=(12,5))
plt.suptitle("Main title")#전체타이틀
#plt.subplots_adjust()을 사용하여 서브플롯간의 간격 변경
plt.subplots_adjust(hspace=0.4, wspace=0.3) #hspace은 위아래 여백 wspace 옆여백
axes[0,0].scatter(x,y,marker='.')
axes[0,0].set(title='An Example Axes',ylabel="y축-axis",xlabel='x축-axis',
             xlim=[0,12],ylim=[-2,2]) #x측 변경
axes[0,1].scatter(x,y,marker='*',c='k')
axes[0,1].set_xlim(2,6)
axes[0,1].set_xlabel('x축 0~5')#눈금의 범위
axes[0,1].set_ylim(-3,3)
axes[0,1].set_ylabel('y축 -3~3')
axes[1,0].scatter(x,y,marker='^',c='r')
#x축 눈금 
axes[1,0].set_xticks(range(0,11,3))#눈금의 간격
axes[1,0].set_xticklabels([-100,3,'6배수','foo'])#눈금의 표시된 값 변경
axes[1,0].set_yticks([-2,0,1,2])
axes[1,0].set_yticklabels(['-2개','제로',1,'Max'])

axes[1,1].scatter(x,y,marker='^',c='y')
axes[1,1].set(xticks=[0,2.5,5,7.5,10],
             xticklabels=[0,2.5,5,7.5,'Max'],
             title='text')
axes[1,1].grid(True)
axes[1,1].spines["top"].set_visible(False)#윗상단 그래프 숨김
axes[1,1].spines["bottom"].set_position(('outward',10))#하단그래프 공간벌리기 #p33 홈페이지 
plt.show()
#https://matplotlib.org/3.1.0/api/axis_api.html 참조
```

**5) 축 공유**

```python
import numpy as np
x= np.arange(0,10)
y1= 0.5*(x**2)#0.5*x*2
y2= -1 * y1

fig,ax1=plt.subplots()
ax1.plot(x,y1,'g^:')
ax1.set(xlabel='Xdata',ylabel='Y1 data')
#ax1.plot([0,1,2,3,4],[0,-10,-20,-30,-40])

ax2=ax1.twinx() #x축을 공유하는 ax2
ax2.plot(x,y2,'b--')
ax2.set_ylabel('y2 data',color='b')

ax3=ax1.twiny() #ax1과 x축을 공유하는 ax3
ax3.plot(-x,y1,'ro-')
ax3.set_xlabel('-x data',color='r')

plt.show()
```
