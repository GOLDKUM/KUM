```python
import requests
from bs4 import BeautifulSoup
url='https://www.melon.com/chart/index.htm' # 크롤링을 막아놓음
headers={'User-agent':'Mozilla/5.0'}
html=requests.get(url,headers=headers)
soup = BeautifulSoup(html.content,'html.parser')
titles=soup.select("div.wrap_song_info > div.ellipsis.rank01 > span > a")
singers=soup.select("div.ellipsis.rank02 > span")
len(titles),len(singers)
for idx in range(len(singers)):
    print("{:3d}위 | {} | {}".format(idx+1,titles[idx].text,singers[idx].text))
```

```python
import requests
from bs4 import BeautifulSoup
```

## 크롤링 가능 홈페이지

```python
url='https://nedrug.mfds.go.kr/searchDrug?sort=&sortOrder=&searchYn=&page=1'
response=requests.get(url)
print(response.json)
soup=BeautifulSoup(response.content,"html.parser")
```

```python
medicine_data=soup.select('table.dr_table.dr_table_type2 tbody td.al_l span a')# 네비게이션홈페이지 리스트 크롤링
list_address_medicine=[]
for i in range(len(medicine_data)):
    list_address_medicine.append('https://nedrug.mfds.go.kr'+medicine_data[i].attrs['href'])
    
```

```python
for j in range(100+1):
    url='https://nedrug.mfds.go.kr/searchDrug?sort=&sortOrder=&searchYn=&page='+str(j)
    response=requests.get(url)
    soup=BeautifulSoup(response.content,"html.parser")
    medicine_data=soup.select('table.dr_table.dr_table_type2 tbody td.al_l span a')
    for i in range(len(medicine_data)):
        list_address_medicine.append('https://nedrug.mfds.go.kr'+medicine_data[i].attrs['href'])
```

```python
effect_list=[]
name_list=[]
for i in range(100+1):
    url=list_address_medicine[i]
    response=requests.get(url)
    soup=BeautifulSoup(response.content,"html.parser")
    name=soup.select('table.s-dr_table span.pb10')
    name_list.append(name[0].string)
    effect=soup.find('div',id='_ee_doc').text.strip()
    effect_list.append(effect)
    
```

```python
import pandas as pd 
df=pd.DataFrame({'제품명':name_list,'효능효과':effect_list})
df
```
