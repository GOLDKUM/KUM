```python
#간단한 점그림
plt.plot(x,np.sin(x),'-og')
plt.plot(x,np.cos(x),
        '-o',
        color='gray',#실선과 원형 마커
        markersize=10,#선의 색은 회색
        linewidth=3,#선의두께
        markerfacecolor='white',#마커의색
        markeredgecolor='gray',#마커테두리
        markeredgewidth=1.5) #마커테두리 두꼐)
plt.ylim(-1.2,1.2)
```

```python
rng=np.random.RandomState(seed=0)
data=rng.randn(50,2)
x,y=data[:,0],data[:,1]
colors=rng.rand(50)
sizes=1000*rng.rand(50)
plt.subplot(2,1,1)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.4,cmap='plasma')
plt.colorbar()
plt.subplot(2,1,2)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.4,cmap='viridis')
plt.colorbar()
```

# subplot 다중 서브 플롯

```python
#plt.subplot
%matplotlib inline
plt.style.use('seaborn-whitegrid')

#조절
fig=plt.figure()
fig.subplots_adjust(hspace=0.4,wspace=0.4)

norws,ncols=3,4
for i in range(norws*ncols):
    plt.subplot(norws,ncols,i+1)
    plt.text(0.5,0.5,str((norws,ncols,i+1)),fontsize=16,ha='center')
    
```

```python
nrows,ncols=3,4
fig,ax=plt.subplots(nrows,ncols,sharex='col',sharey='row')
for i in range(nrows):
    for j in range(ncols):
        ax[i,j].text(0.5,0.5,str((i,j)),fontsize=15,ha='center',color='g')
```

```python
nrows,ncols=3,4
grid=plt.GridSpec(3,4,wspace=0.4,hspace=0.4)
ax1=plt.subplot(grid[0,0])
ax2=plt.subplot(grid[0,1:2+1])
ax3=plt.subplot(grid[1:,:2])
ax4=plt.subplot(grid[1:,2])
ax5=plt.subplot(grid[:,3])

ax1.text(0.5,0.5,'Fig1',fontsize=15,ha='center',color='k')
ax2.text(0.5,0.5,'Fig2',fontsize=15,ha='center',color='k')
ax3.text(0.5,0.5,'Fig3',fontsize=15,ha='center',color='k')
ax4.text(0.5,0.5,'Fig4',fontsize=15,ha='center',color='k')
ax5.text(0.5,0.5,'Fig5',fontsize=15,ha='center',color='k')
```

```python
mean=[0,0]
cov=[[1,1],[1,2]]
x,y=np.random.multivariate_normal(mean,cov,5000).T
```

```python
fig=plt.figure(figsize=(6,6))
grid=plt.GridSpec(4,4,hspace=0.2,wspace=0.2)
main_ax=fig.add_subplot(grid[:-1,1:])
y_hist=fig.add_subplot(grid[:-1,0],xticklabels=[],sharey=main_ax)
x_hist=fig.add_subplot(grid[-1,1:],yticklabels=[],sharey=main_ax)


main_ax.plot(x,y,'ok',markersize=3,alpha=0.2)
x_hist.hist(x,40,histtype='stepfilled',orientation='vertical',color='gray')
x_hist.invert_yaxis()

y_hist.hist(y,40,histtype='stepfilled',orientation='horizontal',color='gray')
y_hist.invert_xaxis()
```
